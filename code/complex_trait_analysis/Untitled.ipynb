```python
annot_file = "/project2/gilad/jpopp/ebQTL/data/ldsc/baselineLD_v2.2_annot/baselineLD.1.annot.gz"
gtex_subsampled_eqtls_loc = "/project2/gilad/jpopp/ebQTL/results/subset_qtl_calling/significant_hits/8pcs/signif_variant_gene_pairs.tsv"
eb_eqtls_loc = "/project2/gilad/jpopp/ebQTL/results/static_qtl_calling/eb_cellid/pseudobulk_tmm/basic/8pcs/mash-signif_variant_gene_pairs.tsv"
```

```python
gtex_eqtls = pd.read_csv(gtex_subsampled_eqtls_loc, sep="\t")
eb_eqtls = pd.read_csv(eb_eqtls_loc, sep="\t")
```

```python
gtex_eqtls_bed = gtex_eqtls[['variant_id', 'tissue']].pivot_table(index='variant_id', columns='tissue', aggfunc=lambda x: 1, fill_value=0).reset_index()
```

How many GTEx eQTLs 

```python
gtex_eqtls_bed
```

```python
gtex_eqtls_bed['CHR'] = [int(s.split('_')[0][3:]) for s in gtex_eqtls_bed['variant_id']]
gtex_eqtls_bed['BP'] = [int(s.split('_')[1]) for s in gtex_eqtls_bed['variant_id']]
gtex_eqtls_bed = gtex_eqtls_bed.drop(columns=['variant_id'])
```

```python
gtex_eqtls_bed
```

Create a new df which is just 1 for all of these GTEx eQTLs

```python
gtexcomb_eqtls_bed = gtex_eqtls_bed[['CHR', 'BP']]
gtexcomb_eqtls_bed['GTEx_eQTL'] = 1
```

```python
gtexcomb_eqtls_bed
```

```python
annots_gtex = annots.merge(gtex_eqtls_bed, on=['CHR', 'BP'], how='left').fillna(0)
```

```python
annots_gtex
```

```python
eb_eqtls
```

```python
eb_eqtls_bed = eb_eqtls[['EB_VARIANT_ID']].drop_duplicates().rename(columns={'EB_VARIANT_ID': 'SNP'})
eb_eqtls_bed['EB'] = 1
```

```python
annots_gtex_eb = annots_gtex.merge(eb_eqtls_bed, on=['SNP'], how='left').fillna(0)
annots_gtex_eb.iloc[:, -50:] = annots_gtex_eb.iloc[:, -50:].astype(int)
```

All-but-one analysis

```python
qtls_combined = annots_gtex_eb.iloc[:, -50:]
```

```python
qtls_combined = qtls_combined.loc[(qtls_combined!=0).any(axis=1)]
```

```python
qtls_unique = 1 - qtls_combined.rename(columns={t: f'Omit_{t}' for t in qtls_combined.columns})
```

```python
qtls_unique
```

```python
annots_leave_one_out = annots.join(qtls_unique).fillna(0)
annots_leave_one_out.iloc[:, -50:] = annots_leave_one_out.iloc[:, -50:].astype(int)
```
